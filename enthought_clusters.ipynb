```python
# Read the CNN

json_file = open('model_1.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
model_1 = model_from_json(loaded_model_json)
model_1.load_weights("model_1.h5")
```

```python
def tree_recognition(im_road, plot_flag=True):
    img_road = load_img(im_road)  # this is a PIL image
    x_road = img_to_array(img_road)
    road_width, road_height, road_depth = np.shape(x_road)
    
    if road_width> road_height:
        size1=3
        size2=20
    else:
        size1=20
        size2=3
    img_width, img_height = 76, 76
    trees_pos = list()
#     Iterate through the image directions
    for i in range(0,road_width-img_width,10):
        for j in range(0,road_height-img_height,10):
            box = (j, i, j+img_width, i+img_height)
            img_ij = img_road.crop(box)
            img_x = img_to_array(img_ij)
            img_x[:,:,0] = img_x[:,:,0]/255
            img_x[:,:,1] = img_x[:,:,1]/255
            img_x[:,:,2] = img_x[:,:,2]/255
            img_x = img_x.reshape(1, img_width, img_height, 3)
            ynew = model_1.predict_classes(img_x)
            if ynew[0][0]==1:
                trees_pos.append([j+img_height/2,i+img_width/2])
    
    name1 = ntpath.basename(im_road)
    if len(trees_pos) != 0:
        trees_pos = np.array(trees_pos)

        # Clustering the trees
        ms = MeanShift(bandwidth=30, bin_seeding=True)
        ms.fit(trees_pos)
        labels = ms.labels_
        cluster_centers = ms.cluster_centers_
        labels_unique = np.unique(labels)
        n_clusters_ = len(labels_unique)
        print("number of trees found : %d" % n_clusters_)
        trees_pos = np.array(trees_pos)
    #     ----------- Plotting the figures -----------
        if plot_flag:
            fig, ax = plt.subplots(figsize=(size1, size2))
            im = plt.imread(im_road)
            implot = plt.imshow(im)
            plt.scatter(x=trees_pos[:,0], y=trees_pos[:,1], c='r', s=40)
            plt.axis('off')
#             plt.savefig('1-' + name1, format='jpg', dpi=1000)
            plt.show()

            fig, ax = plt.subplots(figsize=(size1, size2))
            im = plt.imread(im_road)
            implot = plt.imshow(im)
            plt.scatter(x=cluster_centers[:,0], y=cluster_centers[:,1], c='r', s=40)
            plt.axis('off')
#             plt.savefig('2-' + name1, format='jpg', dpi=1000)
            plt.show()

    return n_clusters_
```

```python
im_road = 'roads/12.jpg'
tree_num = tree_recognition(im_road,plot_flag=True)
```
